# Cleaning: calculating total score

```python
# membuat fungsi yang mengalikan thumbsUpCount dengan score berdasarkan kondisi
def calculate_total_score(row):
    if row['thumbsUpCount'] == 0:
        return row['score']
    elif row['category'] == 'Compliments':
        return row['thumbsUpCount'] * row['score']
    elif row['category'] in ['Critics', 'Complaints']:
        return -row['thumbsUpCount'] * row['score']
    else:
        return row['score']
```

```python
df['totalScore'] = df.apply(calculate_total_score, axis=1)
df.tail()
```

# Cleaning: response time

Code berikut ini dibuat untuk mengetahui response time, yaitu jarak antara review time dengan reply dari pihak RuangGuru

```python
import numpy as np

# menggabungkan antara date dan time untuk waktu review dan reply
df['review_datetime'] = pd.to_datetime(df['review_date'] + ' ' + df['review_time'])
df['reply_datetime'] = pd.to_datetime(df['reply_date'] + ' ' + df['reply_time'])

# menghitung jarak antara review time dengan reply time
df['response_time'] = df['reply_datetime'] - df['review_datetime']

# menghilangkan nilai response_time yang negative menjadi NaT
df['response_time'] = df['response_time'].where(df['response_time'] >= pd.Timedelta(0))

# konversi response_time menjadi bentuk menit
df['response_time_minutes'] = df['response_time'].dt.total_seconds() / 60
```

# Cleaning: Tags
Setelah ditelusuri, value dalam kolom 'tags' masih dalam bentuk string. Untuk menghitung masing-masing tags secara terpisah, maka dapat digunakan library Regex, tujuannya adalah agar string tersebut dapat diolah menjadi bentuk list.

```python
# contoh untuk melihat bentuk value di kolom 'tags' sebelum processing
sample_tags = df['tags'][1]
print(f"Before processing {sample_tags}. dtype: {type(sample_tags)}")
```

```python
# import library Regex
import re

#import numpy
import numpy as np

# mendefinisikan fungsi
def get_tags(data):
    try:
        data = data.strip()
        data = data.split(',')
        tags = [re.findall("\w+", x)[0] for x in data]
        return tags
    except:
      # agar jika nanti ada value NaN, yang di-return adalah nilai NaN kembali.
        return [np.nan]
```

```python
# contoh untuk melihat bentuk value di kolom 'tags' SETELAH processing
# dapat dilihat bahwa dtype telah berubah menjadi list.
example = get_tags(sample_tags)
print(f"After processing: {example}. dtype: {type(example)}")
```

```python
# ekstraksi dari semua tags yang ada di list
all_tags = list()
for tag in df['tags']:
    x = get_tags(tag)
    all_tags += x
```

```python
# menghitung berapa kali masing-masing tags muncul
extracted = pd.Series(all_tags).value_counts()
extracted
```

```python
# menjadikan datanya dalam bentuk dataframe
data = pd.DataFrame({
    'tag':extracted.index,
    'count':extracted.values
})
data
```

```python
def find_avg_tag_rating(data):
    """
    Fungsi untuk menemukan rerata rating dari setiap tag,
    column yang diambil adalah column 'totalScore'
    """
    total_rating = 0
    count = 0
    for i in range(len(df['tags'])):
        try:
            if data in df['tags'][i]:
                total_rating += df['totalScore'][i]
                count += 1
        except:
            continue

    return total_rating / count
```

```python
# ambil satu contoh tags untuk average rating
find_avg_tag_rating('Aplikasi')
```

```python
# apply to the whole dataset
data['rating_avg'] = data['tag'].apply(find_avg_tag_rating)
data
```

```python
# untuk melihat complaints paling banyak apa.
# perlu diingat --> rating_avg bernilai minus = menunjukkan critics dan complaints
data.sort_values(by=['rating_avg'])
```
